```python
def remove_emails(x):
     return re.sub(r'([a-z0-9+._-]+@[a-z0-9+._-]+\.[a-z0-9+_-]+)',"", x)


def remove_urls(x):
    return re.sub(r'(http|https|ftp|ssh)://([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:/~+#-]*[\w@?^=%&/~+#-])?', '' , x)

def remove_rt(x):
    return re.sub(r'\brt\b', '', x).strip()

def remove_special_chars(x):
    x = re.sub(r'[^\w ]+', "", x)
    x = ' '.join(x.split())
    return x
```

```python
train.Text = train.Text.apply(lambda x: remove_emails(x))
train.Text = train.Text.apply(lambda x: remove_urls(x))
train.Text = train.Text.apply(lambda x: remove_rt(x))
train.Text = train.Text.apply(lambda x: remove_special_chars(x))
```

```python
text_features = ['Text']
```

```python
X=train.drop(["Score"],axis=1)
Y=train.Score
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.15,random_state=52)
```

```python
train_pool = Pool(
    x_train,
    y_train,
    text_features=text_features,
    feature_names=list(x_train)
)
valid_pool = Pool(
    x_test,
    y_test,
    text_features=text_features,
    feature_names=list(x_train)
)
```

```python
param = {
    'eval_metric': 'Accuracy'
    }
```

```python
cat=CatBoostClassifier(**param)
```

```python
rs_space={
            'learning_rate': (np.arange(0.01, 0.1, step=0.01)),
            'depth': list(np.arange(1, 10, step=1)),
            'iterations': (np.arange(300, 1000, step=10)),
            'l2_leaf_reg':randint(1,4),
            'random_strength':randint(1,4),
          }
```

```python
cat_random = RandomizedSearchCV(cat, rs_space, n_iter=500, scoring='accuracy', n_jobs=-1, cv=3)
cat_random = cat_random.fit(x_train,y_train)
```

```python
params = {
    'iterations': 200,
    "depth" :5,
    'l2_leaf_reg': 12,
    'learning_rate': 0.8,
    'eval_metric': 'Accuracy',
    'random_strength': 5,
    'use_best_model': True,
    'verbose': 1}
```

```python
cat=CatBoostClassifier(**params)
```

```python
cat.fit(train_pool, eval_set=valid_pool)
```

```python
pred = cat.predict(test)
```

```python
sub.Score=pred
sub.to_csv("su.csv",sep='\t',index=False)
```
